```python
#bibliotheque opencv de traitement d'images
import cv2

#importer une fonctionnalit√© pour afficher la photo
from google.colab.patches import cv2_imshow

#lire l'image et la stocker dans une variable image_hec
image_hec=cv2.imread('/content/sample_data/hec_rabat.jpg')
```

```python
#afficher les couleurs RGB de l'image
print(image_hec)
```

```python
#afficher l'image
cv2_imshow(image_hec)
```

```python
#telecharger directement la photo √† partir d'un site internet
!curl -o hakimi.png 'https://industries.ma/wp-content/uploads/2024/11/achraf-hakimi-696x392.jpg'

#stocker dans la variable hakimi_achraf
hakimi_achraf=cv2.imread('hakimi.png')

#afficher la photo
cv2_imshow(hakimi_achraf)
```

```python
#detecter le visage de hakimi
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gris=cv2.cvtColor(hakimi_achraf,cv2.COLOR_BGR2GRAY)
face=face_cascade.detectMultiScale(gris,1.1,4)

#tracer un rectangle sur le visage trouv√©
for (x,y,w,h) in face:
  cv2.rectangle(hakimi_achraf,(x,y),(x+w,y+h),(255,0,0),2)
  cv2_imshow(hakimi_achraf)
```
