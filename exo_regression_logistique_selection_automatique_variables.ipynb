```python
#importer les donn√©es
import pandas as pd
diabete = pd.read_excel('/content/sample_data/diabete.xlsx')
diabete.head()
```

```python
#recodage de la variable diabete
recodage = {'positive':1,'negative':0}
diabete['cible'] = diabete['diabete'].map(recodage)
diabete.head()
```

```python
#suppression des variables diabete et cible pour garder les variables explicatives
#diabete.drop(['diabete'],axis=1,inplace=True)
diabete.head()
```

```python
diabete['cible'].value_counts()
```

```python
#suppression des variables diabete et cible pour garder les variables explicatives
x = diabete.drop(['diabete','cible'],axis=1)
x.head()
```

```python
#garder la variable cible seule
y = diabete['cible']
y.head()
```

```python
#on cr√©e deux √©chantillons de test avec 20% pour
#l'√©chantillon de test
#les echantillons sont effectu√©s selon un tirage al√©atoire
from sklearn.model_selection import train_test_split
X_app,X_test,y_app,y_test = train_test_split(x,y,test_size=0.2)
print("la base globale est", x.shape)
print("la base d'apprentissage est", X_app.shape)
print("la base de test est", X_test.shape)
```

```python
#importation de la bibliotheque pour la regression logistique
from sklearn.linear_model import LogisticRegression

#calibrage de notre regression logistique
model = LogisticRegression(solver='liblinear')
resultat = model.fit(X_app,y_app)
print(resultat.coef_,resultat.intercept_)
```

```python
print(resultat.score(X_app,y_app))
```

```python
X_app.columns
```

```python
X_test.columns
```

```python
#application du modele sur l'echantillon test
y_pred = resultat.predict(X_test)
#afficher l'√©chantillon test
print(y_pred)
```

```python
#transformation de y_pred en dataframe
df_pred = pd.DataFrame(y_pred)
df_pred.head()
```

```python
df_pred.to_excel('Mesprediction.xlsx')
```

```python
#calcul de la matrice de confusion
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)

print(confusion_matrix(y_test,y_pred))
```

```python
#on stock notre matrice de confusion
matrice = confusion_matrix(y_test,y_pred)
print(matrice)
```

```python
#on stock les valeurs de la matrice de confusion
VP=matrice[0,0]
FP=matrice[0,1]
FN=matrice[1,0]
VN=matrice[1,1]
```

```python
#calcul de l'accuracy
accuracy=(VP+VN)/(VP+VN+FP+FN)
print(accuracy)
```

```python
#on va utiliser la m√©thode RFE de selection des variables
#on va garder uniquemet les variables les plus explicatives
#et qui sont significatives
from sklearn.feature_selection import RFE
selecteur = RFE(estimator=model)

#on stock dans sol le modele avec uniquement les variables explicatives
sol = selecteur.fit(X_app,y_app)

print("Le nombre de variables retenues",sol.n_features_)
print("les variables retenues",sol.support_)
print("classement des variables sorties",sol.ranking_)
```

```python
#afficher les colonnes
X_app.columns
```

```python
#supprimer les colonnes non explicatives
X_new_app=X_app.loc[:,sol.support_]
X_new_app.shape
```

```python
#affichage du nouveau dataframe avec uniquement les variables retenues
#dans le modele finale
X_new_app.head()
```

```python
#estimation du modele finale

modele_final=model.fit(X_new_app,y_app)
print(modele_final.coef_,modele_final.intercept_)
```

```python
#estimer mes nouvelles pr√©visions en fonction du modele
#finale
y_pred_final=modele_final.predict(X_new_app)
print(y_pred_final)
```

```python
df_pred_final=pd.DataFrame(y_pred_final)
df_pred_final.head()
```

```python
df_pred_final.to_excel('MespredictionFinales.xlsx')
```

```python
#la matrice de confusion finale
from sklearn.metrics import confusion_matrix
cm_selection = confusion_matrix(y_app,y_pred_final)
print(cm_selection)
```
