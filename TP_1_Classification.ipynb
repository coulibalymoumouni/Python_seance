```python
#importer les biblioth√®ques de travail
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#charger la dataset diabetes
data_diabetes = pd.read_csv('/content/2021_diabetes.csv',sep=',',index_col=None,header=0)
data_diabetes.head()
```

```python
data_diabetes.info()
```

```python
#Mettons les variables explicatives dans X
X = data_diabetes.drop('Outcome',axis=1)
#Mettons la variable cible dans y
y = data_diabetes['Outcome']
X.head()
```

```python
#Repartir les donn√©es d'entrainement et de test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#importer l'algorithme d'arbe de decision CART
from sklearn.tree import DecisionTreeClassifier
model_TD = DecisionTreeClassifier()
```

```python
#entrainer le mod√®le
model_TD.fit(X_train,y_train)
```

```python
#pr√©diction du mod√®le
y_predTD = model_TD.predict(X_test)
print(y_predTD)
```

```python
#ranger dans un dataframe
df_y_predTD = pd.DataFrame(y_predTD)
df_y_predTD.to_excel("/content/MesprevisDiabetes_2021.xlsx")
```

```python
#evaluer le mod√®le
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
accuracy_TD = accuracy_score(y_test,y_predTD)
print(f"La performance du mod√®le est : {accuracy_TD:.2f}")

precision_TD = precision_score(y_test,y_predTD)
print(f"La precision du mod√®le est : {precision_TD:.2f}")
print('Precision : Une augmentation de la pr√©cision est une r√©duction du nombre de faux positifs')

rappel_TD = recall_score(y_test,y_predTD)
print(f"Le rappel du mod√®le est : {rappel_TD:.2f}")
print('Rappel : Une augmentation du rappel diminue la probabilit√© que la classe positive soit manqu√©e')

f1_score_TD = f1_score(y_test,y_predTD)
print(f"Le f1_score du mod√®le est : {f1_score_TD:.2f}")
print('f1 score : Le f1 score est la moyenne harmonique entre la precision et le rappel')
```

```python
#visualiser l'arbre obtenu
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

feature_names = data_diabetes.columns[:8]
target_names = data_diabetes['Outcome'].unique().tolist()

plt.figure(figsize=(15,15))
plot_tree(model_TD, feature_names = feature_names,
          class_names = ['0','1'],
          filled = True, rounded = True)
plt.savefig('tree_visualisations_diabetes.png')
plt.show()
```

```python
#regenerer l'arbre en limitant la profondeur de l'arbre √† 3
plt.figure(figsize=(15,15))
plot_tree(model_TD, feature_names = feature_names,
          class_names = ['0','1'],
          filled = True, rounded = True, max_depth=3)
plt.savefig('tree_visualisations_diabetes_3.png')
plt.show()

```

```python
#visualisation avec graphviz
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus
features_names = data_diabetes.columns[:8]
dot_data = StringIO()

export_graphviz(model_TD, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True,feature_names = features_names,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('diabetes.png')
Image(graph.create_png())
```
