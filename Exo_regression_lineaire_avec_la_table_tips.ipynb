```python
#importer la biblioth√®que pour les graphiques
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()


```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
#cr√©er mod√®le de regression lineaire
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
import statsmodels.api as sm
#mettre les variables ind√©pendantes x
x = tips["total_bill"]
#mettre la variable d√©pendante y
y = tips["tip"]

#cr√©er le mod√®le
model = sm.OLS(y, x)
#entrainer le mod√®le
res = model.fit()
#afficher les r√©sultats
print(res.summary())
```

```python
#mod√®le avec la constante
x = sm.add_constant(x)
model = sm.OLS(y, x)
res = model.fit()
print(res.summary())
```
