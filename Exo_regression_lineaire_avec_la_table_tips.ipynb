```python
#importer la bibliothèque pour les graphiques
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()


```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
#créer modèle de regression lineaire
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
import statsmodels.api as sm
#mettre les variables indépendantes x
x = tips["total_bill"]
#mettre la variable dépendante y
y = tips["tip"]

#créer le modèle
model = sm.OLS(y, x)
#entrainer le modèle
res = model.fit()
#afficher les résultats
print(res.summary())
```

```python
#modèle avec la constante
x = sm.add_constant(x)
model = sm.OLS(y, x)
res = model.fit()
print(res.summary())
```
